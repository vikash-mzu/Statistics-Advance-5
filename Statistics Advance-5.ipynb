```python
Q1. The 95% confidence interval for the sample with a mean of 50 and a standard deviation of 5 is approximately (48.21, 51.79). This means we are 95% confident that the true population mean lies between 48.21 and 51.79 based on the sample data.

Q2. The chi-square goodness of fit test for the M&M color distribution yields a chi-square statistic of 1.0 and a p-value of approximately 0.96. Since the p-value is much higher than the significance level of 0.05, we fail to reject the null hypothesis. This means the observed distribution of M&M colors does not significantly differ from the expected distribution.
Q3. The chi-square test for the contingency table results in a chi-square statistic of 5.83, a p-value of approximately 0.054, and 2 degrees of freedom. Since the p-value is slightly above the significance level of 0.05, we fail to reject the null hypothesis. This suggests that there is no significant association between the variables in the contingency table.
Q4. 95% Confidence Interval for Proportion of Smokers:
Given:
•	Population = 500
•	Smokers = 60
To calculate the confidence interval for the proportion of smokers, you can use the following Python approach:
import numpy as np
import scipy.stats as stats

# Population data
n_population = 500
n_smokers = 60

# Proportion of smokers
p_hat = n_smokers / n_population

# 95% confidence interval
z = stats.norm.ppf(0.975)  # For 95% confidence
std_error = np.sqrt((p_hat * (1 - p_hat)) / n_population)

# Confidence interval
lower_bound = p_hat - z * std_error
upper_bound = p_hat + z * std_error

(lower_bound, upper_bound)
Q5. 90% Confidence Interval for Sample Mean:
Given:
•	Mean = 75
•	Standard deviation = 12
•	Assume sample size n=30n = 30n=30
You can use the formula for the confidence interval of the mean:
mean = 75
std = 12
n = 30

# 90% confidence interval
confidence_level = 0.90
alpha = 1 - confidence_level
z = stats.norm.ppf(1 - alpha / 2)

# Margin of error
margin_of_error = z * (std / np.sqrt(n))

# Confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

(lower_bound, upper_bound)
Q6. Chi-Square Distribution Plot with 10 Degrees of Freedom:
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Degrees of freedom
df = 10

# Create x values
x = np.linspace(0, 30, 1000)

# Plot chi-square distribution
plt.plot(x, chi2.pdf(x, df), label=f'Chi-square (df={df})')

# Shade area for chi-square statistic of 15
x_fill = np.linspace(0, 15, 1000)
plt.fill_between(x_fill, chi2.pdf(x_fill, df), color='gray', alpha=0.5)

plt.title("Chi-square Distribution (df=10)")
plt.xlabel("Chi-square Statistic")
plt.ylabel("Probability Density")
plt.axvline(15, color='red', linestyle='--', label='Chi-square=15')
plt.legend()
plt.show()
Q7. 99% Confidence Interval for Proportion of Coke Preference:
Given:
•	Population = 1000
•	Preferred Coke = 520
You can use the same method as in Q4 but with different values:
n_population = 1000
n_preferred_coke = 520

# Proportion preferring Coke
p_hat = n_preferred_coke / n_population

# 99% confidence interval
z = stats.norm.ppf(0.995)  # For 99% confidence
std_error = np.sqrt((p_hat * (1 - p_hat)) / n_population)

# Confidence interval
lower_bound = p_hat - z * std_error
upper_bound = p_hat + z * std_error

(lower_bound, upper_bound)
Q8. Chi-Square Goodness of Fit for Coin Bias:
Given:
•	Observed tails = 45
•	Expected tails for a fair coin = 50
Perform the chi-square goodness of fit test:
# Observed frequencies (tails and heads)
observed = np.array([45, 55])  # Assuming 45 tails, 55 heads in 100 flips

# Expected frequencies for a fair coin (50-50 distribution)
expected = np.array([50, 50])

# Perform chi-square test
chi2_stat, p_value = stats.chisquare(observed, expected)

(chi2_stat, p_value)
Q9. Chi-Square Test for Independence (Smoking vs Lung Cancer):
Given a contingency table of smoking status and lung cancer diagnosis, you can use:
# Contingency table
contingency_table = np.array([[20, 15], [10, 25], [15, 20]])

# Perform chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

(chi2_stat, p_value, dof)
Q10. Chi-Square Test for Independence (Chocolate Preference vs Country of Origin):
Given a contingency table for U.S. and U.K. chocolate preferences, you can use:
# Contingency table example
contingency_table = np.array([[180, 150, 170], [140, 160, 200]])

# Perform chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

(chi2_stat, p_value, dof)


```
